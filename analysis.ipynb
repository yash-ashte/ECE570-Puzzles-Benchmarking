```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unrewarded_df = pd.read_csv('Blackbox_PPO/Blackbox_PPO_metrics.csv')
rewarded_df = pd.read_csv('Blackbox_PPO/Blackbox_Rewarded_PPO_metrics.csv')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot Training Loss
axes[0, 0].plot(unrewarded_df['timesteps'], unrewarded_df['train_loss'], label='Unrewarded PPO', color='blue')
axes[0, 0].plot(rewarded_df['timesteps'], rewarded_df['train_loss'], label='Rewarded PPO', color='green')
axes[0, 0].set_title('Training Loss')
axes[0, 0].set_xlabel('Timesteps')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot Policy Gradient Loss
axes[0, 1].plot(unrewarded_df['timesteps'], unrewarded_df['train_policy_gradient_loss'], label='Unrewarded PPO', color='blue')
axes[0, 1].plot(rewarded_df['timesteps'], rewarded_df['train_policy_gradient_loss'], label='Rewarded PPO', color='green')
axes[0, 1].set_title('Policy Gradient Loss')
axes[0, 1].set_xlabel('Timesteps')
axes[0, 1].set_ylabel('Loss')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot Train Number of Updates
axes[1, 0].plot(unrewarded_df['timesteps'], unrewarded_df['train_n_updates'], label='Unrewarded PPO', color='blue')
axes[1, 0].plot(rewarded_df['timesteps'], rewarded_df['train_n_updates'], label='Rewarded PPO', color='green')
axes[1, 0].set_title('Training Updates')
axes[1, 0].set_xlabel('Timesteps')
axes[1, 0].set_ylabel('Number of Updates')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot Value Loss
axes[1, 1].plot(unrewarded_df['timesteps'], unrewarded_df['train_value_loss'], label='Unrewarded PPO', color='blue')
axes[1, 1].plot(rewarded_df['timesteps'], rewarded_df['train_value_loss'], label='Rewarded PPO', color='green')
axes[1, 1].set_title('Value Loss')
axes[1, 1].set_xlabel('Timesteps')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].legend()
axes[1, 1].grid(True)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
rollout_ep_len_mean = (unrewarded_df['rollout_ep_len_mean']).mean()
rollout_ep_len_mean_rewarded = (rewarded_df['rollout_ep_len_mean']).mean()
print(f"rollout_ep_len_mean (Unrewarded): {rollout_ep_len_mean}")
print(f"rollout_ep_len_mean (Rewarded): {rollout_ep_len_mean_rewarded}")
```
